```python
stemmer = PorterStemmer()

def preprocess1(text):
    if isinstance(text, str): # Check if the text is a string
        stop_words = set(stopwords.words("english"))
        text = text.lower()
        text = ''.join([word for word in text if word not in string.punctuation])
        tokens = word_tokenize(text)
        tokens = [word for word in tokens if word not in stop_words]
    #     stemmed_tokens = [stemmer.stem(word) for word in tokens]
        return ' '.join(tokens)
    else: # Handle the case when it's not a string (e.g., NaN)
        return "" # or any other suitable default value
translated_dataset = pd.read_csv('/content/Copy of codemixed roman hindi - Sheet1.csv')
cleaned_text = []
for i in range(translated_dataset.shape[0]):
  sentence = preprocess1(translated_dataset.loc[i].at['text'])
  cleaned_text.append(sentence)
translated_dataset['clean_text'] = cleaned_text
translated_dataset = translated_dataset.drop(['text'],axis = 1)
translated_dataset.rename(columns = {'clean_text':'text'}, inplace = True)
```

```python
translated_dataset = translated_dataset.iloc[:100]
```

```python
max_words = 1000
vectorizer = TfidfVectorizer(max_features = 1000)
vectorizer.fit(translated_dataset['text'])
X_tf = vectorizer.transform(translated_dataset['text'])
X_tf = X_tf.toarray()
print(X_tf.shape)
X_train, X_test, y_train, y_test = train_test_split(X_tf, translated_dataset['label'], test_size=0.2, random_state=42)
```

```python
classifier0 = RandomForestClassifier(n_estimators = 50)
classifier0.fit(X_train, y_train)
print(classifier0.score(X_train, y_train))
print(classifier0.score(X_test, y_test))
y_pred0 = classifier0.predict(X_tf)

classifier1 = DecisionTreeClassifier(max_depth = 50)
classifier1.fit(X_train, y_train)
print(classifier1.score(X_train, y_train))
print(classifier1.score(X_test, y_test))
y_pred0 = classifier1.predict(X_tf)

classifier2 = LogisticRegression()
classifier2.fit(X_train, y_train)
print(classifier2.score(X_train, y_train))
print(classifier2.score(X_test, y_test))
y_pred0 = classifier2.predict(X_tf)

classifier3 = KNeighborsClassifier()
classifier3.fit(X_train, y_train)
print(classifier3.score(X_train, y_train))
print(classifier3.score(X_test, y_test))
y_pred3 = classifier3.predict(X_tf)

classifier4 = GaussianNB()
classifier4.fit(X_train, y_train)
print(classifier4.score(X_train, y_train))
print(classifier4.score(X_test, y_test))
y_pred4 = classifier4.predict(X_tf)

classifier5 = SVC(kernel='rbf')
classifier5.fit(X_train, y_train)
print(classifier5.score(X_train, y_train))
print(classifier5.score(X_test, y_test))
y_pred5 = classifier5.predict(X_tf)

classifier6 = XGBClassifier()
classifier6.fit(X_train, y_train)
print(classifier6.score(X_train, y_train))
print(classifier6.score(X_test, y_test))
y_pred6 = classifier6.predict(X_tf)

classifier7 = GradientBoostingClassifier()
classifier7.fit(X_train, y_train)
print(classifier7.score(X_train, y_train))
print(classifier7.score(X_test, y_test))


classifier8 = CatBoostClassifier()
classifier8.fit(X_train, y_train)
print(classifier8.score(X_train, y_train))
print(classifier8.score(X_test, y_test))

classifier9 = AdaBoostClassifier()
classifier9.fit(X_train, y_train)
print(classifier9.score(X_train, y_train))
print(classifier9.score(X_test, y_test))
```

```python
# Define the GRU model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),  # Input shape is equal to the number of features
    tf.keras.layers.Reshape((1, X_train.shape[1])),  # Reshape to (batch_size, 1, number of features)
    tf.keras.layers.GRU(128),  # GRU layer
    tf.keras.layers.Dense(64, activation='tanh'),  # Dense layer with ReLU activation
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer with sigmoid activation
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=30, batch_size=32)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```
